# Temperature Variable

The temperature variable gives users control over how creative the LLM is in its queries and answers. 

- **Temperature = 1**: The LLM takes more risks and provides more creative responses.
- **Temperature = 0.1 or 0**: The LLM takes minimal to no risks and doesn't attempt to be creative—it provides more deterministic and consistent answers.

```python
llm = OpenAI(openai_api_key= os.environ["OPEN_API_KEY"], temperature=0.6)
```

```python
text = "what is capital of usa "

print(llm.invoke(text))
```

### Combing multiple chains using simple sequential chain 

```python
capital_prompt= PromptTemplate(input_variable=['country'], template="please tell me the capital of the {country}")

capital_chain=LLMChain(llm=llm,prompt=capital_prompt)
```

```python
famous_template = PromptTemplate(input_variable=['capital'], template="what are the some famous places in {capital}")
famous_chain=LLMChain(llm=llm,prompt=famous_template)
```

```python
from langchain.chains import SimpleSequentialChain
chain=SimpleSequentialChain(chains=[capital_chain,famous_chain])
output = chain.run("USA")
print(output)
```

### Sequential chain

```python
capital_prompt= PromptTemplate(input_variable=['country'], template="please tell me the capital of the {country}")

capital_chain=LLMChain(llm=llm,prompt=capital_prompt, output_key="capital")
```

```python
famous_template = PromptTemplate(input_variable=['capital'], template="what are the some famous places in {capital}")
famous_chain=LLMChain(llm=llm,prompt=famous_template, output_key="places")
```

```python
from langchain.chains import SequentialChain 
```

```python
chain = SequentialChain(chains=[capital_chain, famous_chain], input_variables=['country'], output_variables=['capital', 'places'])
```

```python
chain({"country": "USA"})
```

### Prompt Template + LLM + Output Parsers

```python
from langchain.chat_models import ChatOpenAI 
from langchain.prompts.chat import ChatPromptTemplate
from langchain.schema import BaseOutputParser 
```

```python
class Commaseperatdoutput(BaseOutputParser):
    def parse(self,text:str):
        return text.strip().split(",")
```

```python
template = "Your are a helpful assistant. when the user give any input , you should give 3 words as output and those words should be comma seperated list. {text}"
human_template = "{text}"
chatprompt= ChatPromptTemplate.from_messages([
    ("system", template),
    ("human", human_template)
])
```

```python
chain = LLMChain(llm=chatllm, prompt=chatprompt, output_parser=Commaseperatdoutput())
chain.invoke({"text":"hi"})
```
